Extracting only required columns from the data as per Documentation:

price, age_08_04, KM, HP, cc, Doors, Gears, Quarterly_Tax, Weight.

```python
data = df[['Price', 'Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Gears', 'Quarterly_Tax', 'Weight']]
```

```python
data.head()
```

```python
data.columns
```

```python
x = data.drop('Price', axis = 'columns')
y = data['Price']
```

```python
sc = StandardScaler() 
```

```python
X = sc.fit_transform(x)
```

```python
X
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size = 0.25)
```

```python
def model_eval(true, predicted):
    mse = mean_squared_error(true, predicted)
    mae = mean_absolute_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    return mse, mae, rmse
```

```python
models = {
    'Linear Regressor': LinearRegression(),
    'DecisionTreeRegressor' : DecisionTreeRegressor(),
    'Random Forest Regressor' : RandomForestRegressor(n_estimators=100),
    'Ada Boost Regressor': AdaBoostRegressor(),
    'XGB Regressor': XGBRegressor()
}

mdl = []
scr = []

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train, y_train)

    pred = model.predict(X_test)

    mse, mae, rmse = model_eval(y_test, pred)


    print('---------------------------------------------------------\n')
    print("{} Model has :".format(list(models.keys())[i]))
    print('Mean Squared Error : {}'.format(mse))
    print('Mean Absolute Errro : {}'.format(mae))
    print('Root Mean Squared Error : {}\n'.format(rmse))

    print('---------------------------------------------------------\n')
```
